```python
def update_params(w, b, x, y, alpha, m):
    """
    Updating the values of W and B using partial differentiation
    """
    del_w = (1/(m))*(w*x + b - y)*x
    del_b = (1/(m))*(w*x + b - y)
    w = w - alpha * del_w
    b = b - alpha * del_b

    return w, b
```
